```
#1.4.4.1 regular plots
n = 256
X = np.linspace(-np.pi, np.pi, n, endpoint=True)
Y = np.sin(2 * X)

plt.plot(X, Y + 1, color='blue', alpha=1.00)
plt.plot(X, Y - 1, color='blue', alpha=1.00)

ones = np.ones(n)
mones = ones * -1

red = np.less_equal(Y - 1, -1)

blue = np.logical_not(red)

plt.fill_between(X, ones, Y+1, alpha=0.5)
plt.fill_between(X, mones, Y-1, where=blue, color='b', alpha=0.25)
plt.fill_between(X, mones, Y-1, where=red, color='r', alpha=0.25)
```

```
#1.4.4.2 scatter plots
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

theta = np.arctan2(X, Y)

ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

plt.scatter(X, Y, alpha=0.4, c=theta)

print(np.min(theta), np.max(theta))
print(np.arctan2(0.2, -2))
```

```
#3 bar plots
n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
print(Y1, "\n\n", Y2)

plt.bar(X, Y1)
plt.bar(X, -Y2)

for x,y in zip(X, Y1):
    plt.text(x, y, np.round(y, 2))
    
for x,y in zip(X, Y2):
    plt.text(x, -y, np.round(y, 2))
    
round(0.35343784893, 2)
np.round(Y1, 2)
```

```
#4 contour plot
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x, y)

plt.contourf(X, Y, f(X, Y), alpha=0.7)
C = plt.contour(X, Y, f(X, Y), 7, colors="black")
plt.clabel(C)
```
